Lectura de Datos: Crea un DataFrame que contenga los datos provistos en el archivo Datos_Ventas_Tienda.csv provisto en esta lección. El archivo incluirá información como fecha de venta, categoría de producto, cantidad vendida y precio.

```python
import pandas as pd

df1 = pd.read_csv(r"C:\Users\Juan Mares\Jupyter noteoboks\data\Datos_Ventas_Tienda.csv")
df1
```

#### Fusión de Datos: 
Crea un segundo DataFrame que contenga los datos del archivo Datos_Ventas_Tienda2.csv (también provisto en esta lección), y concaténalos para tener un solo dataFrame con toda la información.

```python
df2 = pd.read_csv(r"C:\Users\Juan Mares\Jupyter noteoboks\data\Datos_Ventas_Tienda2.csv")
# Concatenar los 2 DataFrames
df_ventas = pd.concat([df1, df2], ignore_index = True)
df_ventas
```

#### Tratamiento de Datos:
Utiliza Pandas para manipular estos datos. Deberás realizar tareas como limpieza de datos, filtrado y transformaciones básicas.

```python
# Convertir la columna fecha (str) a tipo datetime
df_ventas.info()
```

```python
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])
df_ventas.info()
```

```python
# Verificar que no haya datos nulos
df_ventas.isnull().sum()
```

#### Análisis de Ventas: Realiza análisis para responder preguntas como:

¿Cuál es el producto más vendido?
¿Cuál es el mes con más ventas?

```python
# Producto mas vendido
df_ventas
```

```python
producto_mas_vendido = df_ventas.groupby('Producto')['Cantidad'].sum().sort_values(ascending = False)
producto_mas_vendido
```

```python
# Mayor mes de ventas
meses = []
for f in df_ventas['Fecha']:
    meses.append(f.month)
df_ventas['Meses'] = meses
df_ventas
```

```python
ventas_por_mes = df_ventas.groupby('Meses')['Total Venta'].sum().sort_values(ascending = False)
ventas_por_mes
```

### Datos Agrupados: 
Agrupa los datos por categoría de producto y analiza las ventas por categoría.
Guardar Resultados: Al final, guarda el DataFrame completo (incluyendo la columna de meses) en un archivo .csv en tu ordenador. 

```python
datos = df_ventas.groupby('Producto')['Total Venta'].sum().sort_values(ascending = False)
datos
```
